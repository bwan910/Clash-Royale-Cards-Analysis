```python
# Group by Type = Spawners
spawners = df[df['Type'] == "Spawners"]

# To remove specific columns due to zero value
remove = spawners.drop(['Count','Crown Tower Damage','Damage','Damage per second','Death Damage','Health (+Shield)','Hit Speed','Level','Radius','Range'], axis = 1) 

# To sort by level
cost = remove.sort_values('Cost',ascending=True)
cost.head(100)

# Export spawner dataframe to image
#cost.dfi.export('Spawners.png')
```

```python
# Group by Type = Damage Spells
spells = df[df['Type'] == "Damaging Spells"]

# To remove specific columns due to zero value
remove = spells.drop(['Card Level (Spawn Level)','Count','Damage per second','Death Damage','Health (+Shield)','Hit Speed','Maximum Spawned','Range','Spawn DPS','Spawn Damage','Spawn Health','Spawn Speed','Spawner Health','Troop Spawned'], axis = 1) 

# To sort by level
level= remove.sort_values('Level',ascending=True)
level.head(100)

# Export spells dataframe to image
#level.dfi.export('Damaging Spell.png')
```
